# Insurance Regression Analysis and Insights

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

#Reading datasets from csv file
df = pd.read_csv("C:\\Users\\oliva\\Desktop\\FTW\\FOR_DAY7\\insurance_regression (1).csv")
```

**Count per gender** 

Based from the data analysis, "male" has the highest number of insured compared to "female" 

```python
#describing the count per gender
df.pivot_table(index="sex",values=["age"],aggfunc= "count")
```

```python
df.pivot_table(index="sex",values=["age"],aggfunc= "count").plot.bar(figsize=(8, 8))
plt.title("Number of Insured by Gender")
```

**Count of smokers per gender**

Dissecting the numbers, insured "male" has the highest number of smokers compared to insured "female" smokers.

```python
#describing the count of smokers and non-smokers per gender 
df.pivot_table(index=["sex"],columns=["smoker"],values=["age"],aggfunc= "count")
```

```python
df.pivot_table(index=["sex"],columns=["smoker"],values=["age"],aggfunc= "count").plot.bar(figsize=(8, 8))
plt.title("Number of Smokers by Gender")
```

**Breakdown information Table**

Describing the mean per gender and breakdown into smoker/non-smoker

```python
print("Mean difference between Smoker and Non-Smoker")
print("")
print("Mean - Non-Smoker")
print("")
print(df[df["smoker"]=="no"].mean())
print("\n--------------------------------------------------------------------")
print("")
print("Mean - Smoker")
print("")
print(df[df["smoker"]=="yes"].mean())
```

```python
#Filtering female smokers/non-smokers

female_non_smoker = df[(df["smoker"] == "no") & (df["sex"] == "female")]
female_smoker = df[(df["smoker"] == "yes") & (df["sex"] == "female")]
print("Mean Female Non-Smokers")
print("")
print(female_non_smoker.mean())
print("\n--------------------------------------------------------------------")
print("Mean Female Smokers")
print("")
print(female_smoker.mean())
```

```python
#Filtering male smokers/non-smokers

male_non_smoker = df[(df["smoker"] == "no") & (df["sex"] == "male")]
male_smoker = df[(df["smoker"] == "yes") & (df["sex"] == "male")]
print("Mean Male Non-Smokers")
print("")
print(male_non_smoker.mean())
print("\n--------------------------------------------------------------------")
print("Mean Male Smokers")
print("")
print(male_smoker.mean())
```

**Correlation Analysis**

Based from data analysis, Age do have an effect to the insurance charging cost. As you aged older, the higher the insurance cost will be.  

```python
#Correlation of age to charges

df[["age","charges"]].plot.scatter(x="age",y="charges",figsize=(8, 8))
plt.title("Age vs Charges")
```

It is also perceivable that "smoking" do have an effect to insurance cost.

```python
#describing the difference of insurance cost between smokers and non-smokers.
df.pivot_table(index=["age"],columns=["smoker"],values = "charges",aggfunc= "mean").plot.line(figsize=(8, 8))
plt.title("Smoker vs Non-Smoker")
```

```python
#Correlation per variables

scatter_matrix(df,figsize=(8, 8),color = "orange")
```
